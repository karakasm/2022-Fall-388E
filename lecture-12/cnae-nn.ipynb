```python
import pandas as pd
import numpy as np
import tensorflow as tf

from keras.preprocessing import sequence 
from keras.datasets import imdb

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv1D, Embedding, Dropout

from sklearn.preprocessing import StandardScaler, LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00233/CNAE-9.data',
                   header=None)
data
```

```python
binarizer = LabelBinarizer()
y = binarizer.fit_transform(data.to_numpy()[:,0])
X = data.to_numpy()[:,1:]
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.5)
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(256, input_shape=(X_train.shape[1],), activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(9, activation='sigmoid'))
model.compile(loss = 'categorical_crossentropy', 
              optimizer = 'adam', 
              metrics = ['binary_accuracy'])
```

```python
model.fit(X_train, y_train,
          batch_size = 8,
          epochs = 15,
          validation_data = (X_test, y_test))
```

```python
y_pred = model.predict(X_test).argmax(axis=1)
yy_test = y_test.argmax(axis=1)
confusion_matrix(yy_test,y_pred)
```

```python
print(classification_report(yy_test,y_pred))
```
